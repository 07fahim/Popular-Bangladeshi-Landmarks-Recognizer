```python
version = 2
best_model_name = "vgg19_bn"
models_path = Path("models")
learner = load_learner(f"{best_model_name}_v{version}.pkl")
```

```python
img_path = Path("/content/drive/MyDrive/bangladeshi_landmarks/Copy of dc5d4bb0-580b-40dd-9264-599048ac5eba.webp")

pred_class, pred_idx, probs = learner.predict(img_path)

print(f" Predicted class: {pred_class}")
print(f" Probabilities: {probs[pred_idx]:.4f} (confidence)")
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(img_path)
plt.imshow(img)
plt.axis("off")
plt.title(f"Prediction: {pred_class} ({probs[pred_idx]:.2f})")
plt.show()
```
